```python
import pandas as pd

# Load the uploaded dataset
df = pd.read_csv('ecommerce_customer_data_custom_ratios.csv')  # Replace with actual file name
df.head()


df.info()
df.describe()
df.isnull().sum()
df.columns

# Drop missing rows (optional - only if too many missing values)
df = df.dropna()

# Remove duplicates
df = df.drop_duplicates()

# Check again
df.info()

# Count values in a column
df['Gender'].value_counts()

# Group by age and see purchase
df.groupby('Age')['Total Purchase Amount'].mean()

import matplotlib.pyplot as plt
import seaborn as sns

# Gender distribution
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()

# Purchase by Age
sns.barplot(x='Age', y='Total Purchase Amount', data=df)
plt.title('Average Purchase by Age')
plt.show()
```

I've corrected the column name in the code to `Total Purchase Amount`.

Please run the cell to see the updated plot.

## 📝 Conclusion:

- Most customers are male.
- Age group 25-34 spends the most.
- Electronics is the top-selling category.
